```python
# BLOCCO 4 — Funzione di normalizzazione Z-score corretta
import numpy as np

def normalize_train_test(X_train, X_test):
    """
    Calcola media e deviazione standard SOLO sul training set,
    poi applica la stessa trasformazione al test set.
    Questo evita il data leakage dalla normalizzazione.
    """
    mean = X_train.mean(axis=0)
    std  = X_train.std(axis=0)

    # Evita divisione per zero (metaboliti con varianza nulla)
    std[std == 0] = 1

    X_train_norm = (X_train - mean) / std
    X_test_norm  = (X_test  - mean) / std

    return X_train_norm, X_test_norm, mean, std

print("✅ Funzione di normalizzazione definita correttamente")
```

```python
# BLOCCO 5 — Architettura del CVAE in PyTorch

import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np

# Impostazione seed globale per riproducibilità
SEED = 42
torch.manual_seed(SEED)
np.random.seed(SEED)

# ── Iperparametri dell'architettura ──────────────────────────────────────────
INPUT_DIM   = X.shape[1]  # si aggiorna automaticamente
LATENT_DIM  = 4        # dimensione spazio latente (ridotta per dataset piccolo)
HIDDEN_DIM  = 32       # neuroni per strato nascosto
NUM_CLASSES = 2        # HC e MCI
BETA        = 1.0      # peso del termine KL nella loss (β-VAE)
EPOCHS      = 300      # numero di epoche di training
LR          = 1e-3     # learning rate Adam
N_SYNTHETIC = 30       # campioni sintetici per classe ad ogni fold

# ── Encoder ──────────────────────────────────────────────────────────────────
class CVAEEncoder(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(INPUT_DIM + NUM_CLASSES, HIDDEN_DIM),
            nn.ReLU(),
            nn.Linear(HIDDEN_DIM, HIDDEN_DIM),
            nn.ReLU()
        )
        self.fc_mu      = nn.Linear(HIDDEN_DIM, LATENT_DIM)
        self.fc_logvar  = nn.Linear(HIDDEN_DIM, LATENT_DIM)

    def forward(self, x, y_onehot):
        xy = torch.cat([x, y_onehot], dim=1)
        h  = self.net(xy)
        return self.fc_mu(h), self.fc_logvar(h)

# ── Decoder ──────────────────────────────────────────────────────────────────
class CVAEDecoder(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(LATENT_DIM + NUM_CLASSES, HIDDEN_DIM),
            nn.ReLU(),
            nn.Linear(HIDDEN_DIM, HIDDEN_DIM),
            nn.ReLU(),
            nn.Linear(HIDDEN_DIM, INPUT_DIM)
        )

    def forward(self, z, y_onehot):
        zy = torch.cat([z, y_onehot], dim=1)
        return self.net(zy)

# ── CVAE completo ─────────────────────────────────────────────────────────────
class CVAE(nn.Module):
    def __init__(self):
        super().__init__()
        self.encoder = CVAEEncoder()
        self.decoder = CVAEDecoder()

    def reparameterize(self, mu, logvar):
        std = torch.exp(0.5 * logvar)
        eps = torch.randn_like(std)
        return mu + eps * std

    def forward(self, x, y_onehot):
        mu, logvar = self.encoder(x, y_onehot)
        z          = self.reparameterize(mu, logvar)
        x_recon    = self.decoder(z, y_onehot)
        return x_recon, mu, logvar

# ── Loss ELBO (ricostruzione + KL) ───────────────────────────────────────────
def cvae_loss(x_recon, x, mu, logvar, beta=BETA):
    recon_loss = nn.functional.mse_loss(x_recon, x, reduction='sum')
    kl_loss    = -0.5 * torch.sum(1 + logvar - mu.pow(2) - logvar.exp())
    return recon_loss + beta * kl_loss

# ── Funzione di training ──────────────────────────────────────────────────────
def train_cvae(X_train_norm, y_train):
    """
    Addestra il CVAE su X_train_norm e y_train.
    Restituisce il modello addestrato.
    """
    model     = CVAE()
    optimizer = optim.Adam(model.parameters(), lr=LR, weight_decay=1e-4)

    X_tensor = torch.FloatTensor(X_train_norm)
    y_onehot = torch.zeros(len(y_train), NUM_CLASSES)
    y_onehot.scatter_(1, torch.LongTensor(y_train).unsqueeze(1), 1)

    model.train()
    for epoch in range(EPOCHS):
        optimizer.zero_grad()
        x_recon, mu, logvar = model(X_tensor, y_onehot)
        loss = cvae_loss(x_recon, X_tensor, mu, logvar)
        loss.backward()
        optimizer.step()

    return model

# ── Funzione di generazione campioni sintetici ────────────────────────────────
def generate_synthetic(model, n_per_class=N_SYNTHETIC):
    """
    Genera n_per_class campioni sintetici per ciascuna classe (HC e MCI).
    Restituisce X_synthetic e y_synthetic.
    """
    model.eval()
    X_syn_list = []
    y_syn_list = []

    with torch.no_grad():
        for label in [0, 1]:
            y_onehot = torch.zeros(n_per_class, NUM_CLASSES)
            y_onehot[:, label] = 1
            z = torch.randn(n_per_class, LATENT_DIM)
            x_syn = model.decoder(z, y_onehot).numpy()
            X_syn_list.append(x_syn)
            y_syn_list.extend([label] * n_per_class)

    return np.vstack(X_syn_list), np.array(y_syn_list)

print("✅ Architettura CVAE definita correttamente")
print(f"   • Spazio latente: {LATENT_DIM} dimensioni")
print(f"   • Neuroni per strato: {HIDDEN_DIM}")
print(f"   • Epoche di training: {EPOCHS}")
print(f"   • Beta (peso KL): {BETA}")
print(f"   • Campioni sintetici per classe per fold: {N_SYNTHETIC}")
```

```python
# BLOCCO 6 — Loop LOOCV corretto con CVAE annidato

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score
from sklearn.linear_model import LassoCV

torch.manual_seed(SEED)
np.random.seed(SEED)

n_samples    = len(y)
y_true_all   = []
y_prob_all   = []
fold_results = []

print("Avvio LOOCV corretto (14 fold)...\n")

for test_idx in range(n_samples):

    # 1. Split train / test
    train_idx   = [i for i in range(n_samples) if i != test_idx]
    X_train_raw = X[train_idx]
    y_train     = y[train_idx]
    X_test_raw  = X[[test_idx]]
    y_test      = y[test_idx]

    # 2. Normalizzazione solo sul training
    X_train_norm, X_test_norm, _, _ = normalize_train_test(X_train_raw, X_test_raw)

    # 3. Training CVAE solo sui dati di training
    torch.manual_seed(SEED)
    cvae_model = train_cvae(X_train_norm, y_train)

    # 4. Generazione campioni sintetici
    X_syn, y_syn = generate_synthetic(cvae_model, n_per_class=N_SYNTHETIC)

    # 5. Dataset aumentato
    X_aug = np.vstack([X_train_norm, X_syn])
    y_aug = np.concatenate([y_train, y_syn])

    # 6. Feature selection con LASSO sul training aumentato
    lasso     = LassoCV(cv=3, random_state=SEED, max_iter=5000).fit(X_aug, y_aug)
    lasso_mask = lasso.coef_ != 0
    if lasso_mask.sum() == 0:
        lasso_mask = np.ones(X_aug.shape[1], dtype=bool)

    X_aug_sel  = X_aug[:, lasso_mask]
    X_test_sel = X_test_norm[:, lasso_mask]

    # 7. Training Random Forest
    rf = RandomForestClassifier(n_estimators=500, random_state=SEED)
    rf.fit(X_aug_sel, y_aug)

    # 8. Predizione sul test reale
    prob = rf.predict_proba(X_test_sel)[0, 1]
    pred = rf.predict(X_test_sel)[0]

    y_true_all.append(y_test)
    y_prob_all.append(prob)

    stato           = "OK" if pred == y_test else "ERRORE"
    classe_reale    = "MCI" if y_test == 1 else "HC"
    classe_predetta = "MCI" if pred   == 1 else "HC"

    fold_results.append({
        "campione"      : sample_names[test_idx],
        "reale"         : classe_reale,
        "predetto"      : classe_predetta,
        "prob_MCI"      : round(prob, 3),
        "corretto"      : stato,
        "n_feature_sel" : int(lasso_mask.sum())
    })

    print(f"  Fold {test_idx+1:2d} | {sample_names[test_idx]} | Reale: {classe_reale} | Predetto: {classe_predetta} | P(MCI)={prob:.3f} | {stato}")

# Calcolo AUC finale
y_true_all = np.array(y_true_all)
y_prob_all = np.array(y_prob_all)
auc        = roc_auc_score(y_true_all, y_prob_all)
accuracy   = np.mean(y_true_all == (y_prob_all >= 0.5).astype(int))

print("\n" + "="*60)
print("LOOCV completato!")
print(f"  AUC      = {auc:.4f}")
print(f"  Accuracy = {accuracy:.4f}")
print("="*60)
```

```python
# BLOCCO 7 — Metriche complete con intervallo di confidenza (bootstrap)

from sklearn.metrics import (confusion_matrix, classification_report,
                             roc_auc_score, roc_curve)
from scipy import stats

# ── Predizioni binarie al cutoff 0.5 ────────────────────────────────────────
y_pred_all = (y_prob_all >= 0.5).astype(int)

# ── Matrice di confusione ────────────────────────────────────────────────────
cm = confusion_matrix(y_true_all, y_pred_all)
TN, FP, FN, TP = cm.ravel()

sensitivity = TP / (TP + FN)
specificity = TN / (TN + FP)
ppv         = TP / (TP + FP)   # Positive Predictive Value
npv         = TN / (TN + FN)   # Negative Predictive Value
accuracy    = (TP + TN) / (TP + TN + FP + FN)

print("=" * 50)
print("METRICHE DI CLASSIFICAZIONE (LOOCV)")
print("=" * 50)
print(f"  TP={TP}  TN={TN}  FP={FP}  FN={FN}")
print(f"  Sensitivity (Recall) : {sensitivity:.4f}")
print(f"  Specificity          : {specificity:.4f}")
print(f"  PPV (Precision)      : {ppv:.4f}")
print(f"  NPV                  : {npv:.4f}")
print(f"  Accuracy             : {accuracy:.4f}")
print(f"  AUC                  : {roc_auc_score(y_true_all, y_prob_all):.4f}")

# ── Intervallo di confidenza AUC via bootstrap (2000 iterazioni) ─────────────
np.random.seed(SEED)
n_boot    = 2000
auc_boots = []

for _ in range(n_boot):
    idx = np.random.choice(len(y_true_all), len(y_true_all), replace=True)
    if len(np.unique(y_true_all[idx])) < 2:
        continue
    auc_boots.append(roc_auc_score(y_true_all[idx], y_prob_all[idx]))

auc_ci_low  = np.percentile(auc_boots, 2.5)
auc_ci_high = np.percentile(auc_boots, 97.5)

print(f"\n  AUC 95% CI (bootstrap, n=2000): [{auc_ci_low:.4f} – {auc_ci_high:.4f}]")

# ── Tabella riassuntiva fold per fold ────────────────────────────────────────
print("\n" + "=" * 50)
print("DETTAGLIO PER CAMPIONE")
print("=" * 50)
df_results = pd.DataFrame(fold_results)
display(df_results)
```

```python
# BLOCCO 7b — Test di permutazione sull'AUC (versione ottimizzata)
import matplotlib.pyplot as plt
np.random.seed(SEED)

N_PERMUTATIONS = 200   # ridotto per velocità
auc_permuted   = []

# Versione veloce: CVAE con meno epoche per le permutazioni
EPOCHS_PERM = 100      # invece di 300

def train_cvae_fast(X_train_norm, y_train):
    model     = CVAE()
    optimizer = optim.Adam(model.parameters(), lr=LR)
    X_tensor  = torch.FloatTensor(X_train_norm)
    y_onehot  = torch.zeros(len(y_train), NUM_CLASSES)
    y_onehot.scatter_(1, torch.LongTensor(y_train).unsqueeze(1), 1)
    model.train()
    for epoch in range(EPOCHS_PERM):
        optimizer.zero_grad()
        x_recon, mu, logvar = model(X_tensor, y_onehot)
        loss = cvae_loss(x_recon, X_tensor, mu, logvar)
        loss.backward()
        optimizer.step()
    return model

print(f"Avvio test di permutazione ({N_PERMUTATIONS} iterazioni)...")
print("Attendi circa 5 minuti...\n")

for perm_i in range(N_PERMUTATIONS):

    y_permuted  = np.random.permutation(y)
    y_prob_perm = []

    for test_idx in range(n_samples):

        train_idx    = [i for i in range(n_samples) if i != test_idx]
        X_train_raw  = X[train_idx]
        y_train_p    = y_permuted[train_idx]
        X_test_raw   = X[[test_idx]]

        X_train_norm, X_test_norm, _, _ = normalize_train_test(X_train_raw, X_test_raw)

        torch.manual_seed(SEED + perm_i)
        cvae_p           = train_cvae_fast(X_train_norm, y_train_p)
        X_syn_p, y_syn_p = generate_synthetic(cvae_p, n_per_class=N_SYNTHETIC)

        X_aug_p = np.vstack([X_train_norm, X_syn_p])
        y_aug_p = np.concatenate([y_train_p, y_syn_p])

        # Feature selection semplificata con soglia fissa
        # per evitare i ConvergenceWarning di LassoCV
        from sklearn.linear_model import Lasso
        lasso_p = Lasso(alpha=0.01, max_iter=2000).fit(X_aug_p, y_aug_p)
        mask_p  = lasso_p.coef_ != 0
        if mask_p.sum() == 0:
            mask_p = np.ones(X_aug_p.shape[1], dtype=bool)

        rf_p = RandomForestClassifier(
            n_estimators=100,   # ridotto per velocità
            random_state=SEED
        )
        rf_p.fit(X_aug_p[:, mask_p], y_aug_p)
        prob_p = rf_p.predict_proba(X_test_norm[:, mask_p])[0, 1]
        y_prob_perm.append(prob_p)

    try:
        auc_p = roc_auc_score(y_permuted, np.array(y_prob_perm))
    except:
        auc_p = 0.5
    auc_permuted.append(auc_p)

    if (perm_i + 1) % 50 == 0:
        print(f"  {perm_i+1}/{N_PERMUTATIONS} completate...")

# ── Risultati ────────────────────────────────────────────────────────────────
auc_observed = roc_auc_score(y_true_all, y_prob_all)
p_value_perm = np.mean(np.array(auc_permuted) >= auc_observed)

print("\n" + "="*55)
print("RISULTATI TEST DI PERMUTAZIONE")
print("="*55)
print(f"  AUC osservata:              {auc_observed:.4f}")
print(f"  AUC media permutazioni:     {np.mean(auc_permuted):.4f}")
print(f"  AUC mediana permutazioni:   {np.median(auc_permuted):.4f}")
print(f"  p-value:                    {p_value_perm:.4f}")
print(f"  Permutazioni:               {N_PERMUTATIONS}")

if p_value_perm < 0.05:
    print(f"\n  L'AUC osservata e' significativamente superiore al caso (p < 0.05)")
else:
    print(f"\n  L'AUC osservata NON raggiunge la significativita' statistica (p >= 0.05)")
print("="*55)

# ── Grafico ──────────────────────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(7, 5), dpi=150)
ax.hist(auc_permuted, bins=25, color='lightgray',
        edgecolor='gray', alpha=0.8, label='Permuted AUC distribution')
ax.axvline(auc_observed, color='steelblue', lw=2.5,
           label=f'Observed AUC = {auc_observed:.3f}')
ax.axvline(np.percentile(auc_permuted, 95), color='red',
           lw=1.5, linestyle='--', label='95th percentile (p = 0.05)')
ax.set_xlabel('AUC', fontsize=12)
ax.set_ylabel('Frequency', fontsize=12)
ax.set_title(
    f'Permutation Test — AUC Distribution\n(p = {p_value_perm:.4f}, n = {N_PERMUTATIONS} permutations)',
    fontsize=12, fontweight='bold')
ax.legend(fontsize=9)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('permutation_test_AUC.png', dpi=300, bbox_inches='tight')
plt.show()
print("Figura salvata: permutation_test_AUC.png")
```

```python
# BLOCCO 8 — Validazione del CVAE: MMD + KS test

from scipy.stats import ks_2samp

# ── Addestriamo UN CVAE sull'intero dataset normalizzato ─────────────────────
# (solo per validazione — NON viene usato per la classificazione)
X_all_norm = (X - X.mean(axis=0)) / (X.std(axis=0) + 1e-8)

torch.manual_seed(SEED)
cvae_val   = train_cvae(X_all_norm, y)
X_syn_val, y_syn_val = generate_synthetic(cvae_val, n_per_class=N_SYNTHETIC)

# ── Maximum Mean Discrepancy (MMD) ───────────────────────────────────────────
def mmd_rbf(X, Y, gamma=1.0):
    """
    Calcola la MMD con kernel RBF tra due distribuzioni X e Y.
    Valori vicini a 0 indicano distribuzioni simili.
    """
    from sklearn.metrics.pairwise import rbf_kernel
    XX = rbf_kernel(X, X, gamma)
    YY = rbf_kernel(Y, Y, gamma)
    XY = rbf_kernel(X, Y, gamma)
    return XX.mean() + YY.mean() - 2 * XY.mean()

mmd_value = mmd_rbf(X_all_norm, X_syn_val)

print("=" * 55)
print("VALIDAZIONE CVAE")
print("=" * 55)
print(f"\n  Maximum Mean Discrepancy (MMD, RBF kernel):")
print(f"  MMD = {mmd_value:.6f}  (più vicino a 0 = più simile al reale)")

# ── KS test per ogni metabolita ──────────────────────────────────────────────
print(f"\n  Kolmogorov-Smirnov test per metabolita")
print(f"  (p > 0.05 = nessuna differenza significativa)\n")
print(f"  {'Metabolita':<35} {'KS stat':>8}  {'p-value':>10}  {'Esito':>15}")
print(f"  {'-'*75}")

ks_stats  = []
ks_pvals  = []
n_ok      = 0

for i, feat in enumerate(feature_names):
    stat, pval = ks_2samp(X_all_norm[:, i], X_syn_val[:, i])
    ks_stats.append(stat)
    ks_pvals.append(pval)
    esito = "p>0.05 (OK)" if pval > 0.05 else "p<=0.05 (diverso)"
    if pval > 0.05:
        n_ok += 1
    print(f"  {feat:<35} {stat:>8.4f}  {pval:>10.4f}  {esito:>15}")

print(f"\n  Metaboliti con p > 0.05: {n_ok}/{len(feature_names)}")
print(f"  MMD = {mmd_value:.6f}")
print("=" * 55)
```

```python
# BLOCCO 9 — Feature importance consensuale sull'intero dataset aumentato

from xgboost import XGBClassifier
from sklearn.preprocessing import MinMaxScaler

# ── Dataset aumentato sull'intero dataset normalizzato ───────────────────────
# (usato SOLO per feature ranking — non per valutazione delle performance)
X_all_norm = (X - X.mean(axis=0)) / (X.std(axis=0) + 1e-8)

torch.manual_seed(SEED)
cvae_full        = train_cvae(X_all_norm, y)
X_syn_full, y_syn_full = generate_synthetic(cvae_full, n_per_class=N_SYNTHETIC)

X_aug_full = np.vstack([X_all_norm, X_syn_full])
y_aug_full = np.concatenate([y, y_syn_full])

# ── 1. Random Forest importance ───────────────────────────────────────────────
rf_full = RandomForestClassifier(n_estimators=500, random_state=SEED)
rf_full.fit(X_aug_full, y_aug_full)
imp_rf = rf_full.feature_importances_

# ── 2. XGBoost importance ─────────────────────────────────────────────────────
xgb = XGBClassifier(n_estimators=500, random_state=SEED,
                    use_label_encoder=False, eval_metric='logloss',
                    verbosity=0)
xgb.fit(X_aug_full, y_aug_full)
imp_xgb = xgb.feature_importances_

# ── 3. LASSO coefficienti ────────────────────────────────────────────────────
lasso_full = LassoCV(cv=3, random_state=SEED, max_iter=5000)
lasso_full.fit(X_aug_full, y_aug_full)
imp_lasso = np.abs(lasso_full.coef_)

# ── Normalizzazione 0-1 per ogni metodo ──────────────────────────────────────
def norm01(v):
    r = v - v.min()
    return r / r.max() if r.max() > 0 else r

imp_rf_n    = norm01(imp_rf)
imp_xgb_n   = norm01(imp_xgb)
imp_lasso_n = norm01(imp_lasso)

# ── Consensus score = media delle tre importanze normalizzate ────────────────
consensus = (imp_rf_n + imp_xgb_n + imp_lasso_n) / 3

# ── Tabella ordinata per consensus score ─────────────────────────────────────
df_imp = pd.DataFrame({
    "Metabolita"  : feature_names,
    "RF"          : imp_rf_n.round(4),
    "XGBoost"     : imp_xgb_n.round(4),
    "LASSO"       : imp_lasso_n.round(4),
    "CONSENSUS"   : consensus.round(4)
}).sort_values("CONSENSUS", ascending=False).reset_index(drop=True)

df_imp.index += 1   # indice parte da 1

print("=" * 70)
print("FEATURE IMPORTANCE CONSENSUALE (top 10)")
print("=" * 70)
display(df_imp.head(10))

print(f"\nSoglia consensus per top 3: {df_imp['CONSENSUS'].iloc[2]:.4f}")
print(f"Soglia consensus per top 5: {df_imp['CONSENSUS'].iloc[4]:.4f}")
```

```python
# BLOCCO BASELINE — RF senza augmentation CVAE
# Confronto diretto con la pipeline principale

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score
from sklearn.linear_model import LassoCV
import numpy as np

np.random.seed(SEED)

y_true_base = []
y_prob_base = []

print("Avvio LOOCV baseline (senza CVAE)...\n")

for test_idx in range(n_samples):

    # 1. Split
    train_idx    = [i for i in range(n_samples) if i != test_idx]
    X_train_raw  = X[train_idx]
    y_train      = y[train_idx]
    X_test_raw   = X[[test_idx]]
    y_test       = y[test_idx]

    # 2. Normalizzazione solo sul training
    X_train_norm, X_test_norm, _, _ = normalize_train_test(X_train_raw, X_test_raw)

    # 3. Feature selection LASSO solo sui dati reali (no sintetici)
    lasso_b = LassoCV(cv=3, random_state=SEED, max_iter=5000).fit(X_train_norm, y_train)
    mask_b  = lasso_b.coef_ != 0
    if mask_b.sum() == 0:
        mask_b = np.ones(X_train_norm.shape[1], dtype=bool)

    # 4. RF solo sui dati reali (no sintetici)
    rf_b = RandomForestClassifier(n_estimators=500, random_state=SEED)
    rf_b.fit(X_train_norm[:, mask_b], y_train)

    # 5. Predizione
    prob_b = rf_b.predict_proba(X_test_norm[:, mask_b])[0, 1]
    pred_b = rf_b.predict(X_test_norm[:, mask_b])[0]

    y_true_base.append(y_test)
    y_prob_base.append(prob_b)

    stato        = "OK" if pred_b == y_test else "ERRORE"
    classe_reale = "MCI" if y_test == 1 else "HC"
    classe_pred  = "MCI" if pred_b == 1 else "HC"
    print(f"  Fold {test_idx+1:2d} | {sample_names[test_idx]} | "
          f"Reale: {classe_reale} | Predetto: {classe_pred} | "
          f"P(MCI)={prob_b:.3f} | {stato}")

# ── Metriche baseline ─────────────────────────────────────────────────────────
y_true_base = np.array(y_true_base)
y_prob_base = np.array(y_prob_base)
y_pred_base = (y_prob_base >= 0.5).astype(int)

auc_base      = roc_auc_score(y_true_base, y_prob_base)
acc_base      = np.mean(y_true_base == y_pred_base)
from sklearn.metrics import confusion_matrix
cm_b          = confusion_matrix(y_true_base, y_pred_base)
tn_b, fp_b, fn_b, tp_b = cm_b.ravel()
sens_base     = tp_b / (tp_b + fn_b)
spec_base     = tn_b / (tn_b + fp_b)

print("\n" + "="*55)
print("CONFRONTO: Baseline vs CVAE-augmented")
print("="*55)
print(f"{'Metrica':<20} {'Baseline (no CVAE)':>20} {'CVAE-augmented':>16}")
print(f"{'-'*55}")
print(f"{'AUC':<20} {auc_base:>20.4f} {0.8333:>16.4f}")
print(f"{'Accuracy':<20} {acc_base:>20.4f} {0.7143:>16.4f}")
print(f"{'Sensitivity':<20} {sens_base:>20.4f} {0.7500:>16.4f}")
print(f"{'Specificity':<20} {spec_base:>20.4f} {0.6667:>16.4f}")
print(f"{'TP':<20} {tp_b:>20} {6:>16}")
print(f"{'TN':<20} {tn_b:>20} {4:>16}")
print(f"{'FP':<20} {fp_b:>20} {2:>16}")
print(f"{'FN':<20} {fn_b:>20} {2:>16}")
print("="*55)
```
